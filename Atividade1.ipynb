```python
# Importações
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Constantes de entrada
P1 = np.repeat(np.random.random(20) * 100 + 101000, 20) # Pa
V1 = np.tile((np.sort(np.random.random(20)) * 5), 20) # m/s
Z1 = 10 # m

P2 = 101000 # Pa
Z2 = 2 # m

RO = 1000 # kg/m³
G = 9.81 # m/s^2
```

```python
# Calculando
V2 = np.sqrt((P1 / RO + V1**2 / 2*G + Z1 - P2 / RO - Z2) * 2*G)
```

```python
# Plotagens
df = pd.DataFrame(
    {
        'Pressão Inicial (Pa)': np.round(P1, 2),
        'Velocidade Inicial (m/s)': np.round(V1, 2),
        'Velocidade Final (m/s)': V2,
        'Cenário': [i // 20 for i in range(20*20)]
    }
)

df
```

```python
# Plotando
g = sns.catplot(
    data=df,
    x='Velocidade Inicial (m/s)',
    y='Velocidade Final (m/s)',
    col='Pressão Inicial (Pa)',
    kind='point',
    markers="o",
    linestyles="--",
    linewidth=2,
    color='#53B2DB',
    col_wrap=2,
)

for ax in g.axes.flat:
    plt.setp(ax.get_xticklabels(), rotation=45, ha="right")
```
